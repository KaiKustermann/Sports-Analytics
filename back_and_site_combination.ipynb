```python
import pandas as pd

landmark_back_df = pd.read_json("./landmark_results/Gehen-6,5_cut.json")
landmark_site_df = pd.read_json("./landmark_results/Gehen-6,5-seite_cut.json")

def adjust_coordinates(landmark_back_df, landmark_site_df):
    for column in landmark_back_df.columns:
        if column[-4:] != "_vis":
            df_back_column_mean = landmark_back_df[column].mean()
            df_site_column_mean = landmark_site_df[column].mean()
            delta_mean = df_back_column_mean - df_site_column_mean
            landmark_site_df[column] += delta_mean
    return landmark_back_df, landmark_site_df

def combine_angles(landmark_back_df, landmark_site_df):
    landmark_combined_df = (landmark_back_df + landmark_site_df) / 2
    return landmark_combined_df

landmark_back_df = adjust_coordinates(landmark_back_df, landmark_site_df)[0]
landmark_site_df = adjust_coordinates(landmark_back_df, landmark_site_df)[1]

landmark_combined_df = combine_angles(landmark_back_df, landmark_site_df)
landmark_combined_df
```
