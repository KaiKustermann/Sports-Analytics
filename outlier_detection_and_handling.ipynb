```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
def plot_features(csv_name):
    landmark_df = pd.read_csv("./landmark_results/{}.csv".format(csv_name), index_col=0)
    for column in landmark_df.columns:   
        plt.plot(landmark_df[column])
        plt.xlabel('frame')
        if landmark_df[column].name[-4:] == "_vis":
            plt.ylabel('visibility')
        else:
            plt.ylabel('landmark_position')
        plt.title(column)
        plt.show()

plot_features("1_GehenH")
```
